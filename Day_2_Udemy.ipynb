###**Number 1**

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load the Iris dataset from seaborn
iris = sns.load_dataset('iris')

# Step 2: Inspect the dataset
# Check the first few rows
print("First five rows of the dataset:")
print(iris.head())

# Check for missing values
print("\nMissing values in the dataset:")
print(iris.isnull().sum())

# Step 3: Basic EDA - Descriptive statistics
print("\nDescriptive statistics:")
print(iris.describe())

# Step 4: Handling missing values
# For demonstration purposes, let's assume some values are missing and handle them
# (Since Iris dataset doesn't have missing values, we artificially add some missing values)

# Introducing missing values in a copy of the dataset
iris_with_missing = iris.copy()
iris_with_missing.iloc[2:5, 0] = None  # Add missing values in 'sepal_length'

# Handling missing values by filling them with the mean of the column
iris_with_missing['sepal_length'].fillna(iris_with_missing['sepal_length'].mean(), inplace=True)

print("\nDataset after filling missing values in 'sepal_length':")
print(iris_with_missing.head())

# Step 5: Visualizing relationships between features
# Pairplot to visualize relationships between all features, colored by species
print("\nVisualizing relationships between features (pairplot):")
sns.pairplot(iris, hue='species')
plt.show()

# Correlation heatmap
print("\nCorrelation heatmap of numerical features:")
numeric_features = iris.select_dtypes(include=['number'])
corr_matrix = numeric_features.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

###**Question 2**

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load the Titanic dataset from seaborn
titanic = sns.load_dataset('titanic')

# Step 2: Count the number of survivors and non-survivors
survival_counts = titanic['survived'].value_counts()

# Step 3: Create labels for the pie chart
labels = ['Did not survive', 'Survived']

# Step 4: Create the pie chart
plt.figure(figsize=(6, 6))
plt.pie(survival_counts, labels=labels, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightblue'])
plt.title('Proportion of Passengers Who Survived vs. Did Not Survive')
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset with income column (replace this with your actual dataset)
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Income': [50000, 60000, 55000, 75000, 68000]
}

# Step 1: Load the dataset into a DataFrame
df = pd.DataFrame(data)

# Step 2: Find the median income
median_income = df['Income'].median()
print(f"The median income is: {median_income}")

# Step 3: Create a histogram of income distribution
plt.figure(figsize=(8, 6))
plt.hist(df['Income'], bins=5, color='skyblue', edgecolor='black')
plt.title('Income Distribution')
plt.xlabel('Income')
plt.ylabel('Frequency')

# Display the histogram
plt.axvline(median_income, color='red', linestyle='dashed', linewidth=1.5, label=f'Median: {median_income}')
plt.legend()
plt.show()
```
